```python
#pandas, openpyxl 라이브러리 부르기
import pandas as pd
import numpy as np
import os
import openpyxl
from tabulate import tabulate
import matplotlib.pyplot as plt


##  데이터 부르기
path = '/content/drive/MyDrive/python/entrance/'
#df = pd.read_excel(f'{path}jeongsi_fouryears.xlsx')

df = pd.read_excel(f'{path}j_new01.xlsx')


# 첫 5줄을 봅니다.
df.head(2)
```

```python
# 연도 입력받기
year = int(input("연도를 입력하세요: "))
college = input("대학을 입력하세요: ")
major = input("전공을 입력하세요: ")

# 해당 연도의 데이터 필터링
filtered_df = df[(df['연도'] == year) & (df['대학'] == college)]

# 전공 필터링
filtered_df = filtered_df[filtered_df['학과'].str.contains(major)]


#필요한 열만 추출
filtered_df = filtered_df[['대학', '연도', '전형', '학과', '모집인원', '실경쟁률']]


#필요한 열만 추출
filtered_df['실경쟁률'] = filtered_df['실경쟁률'].round(1)



# 결과 출력
if not filtered_df.empty:
    print()
    print(f"{year}년도 {college}의 데이터:")
    print(filtered_df)
else:
    print(f"{year}년도 {college} 대학에 해당하는 데이터가 없습니다.")
```

```python
#특정 회사의 특정시기 주가

minds_data = fdr.DataReader("377480", "2022-10-27", "2023-10-27")
display(minds_data)
```

```python
#특정 회사의 특정시기 주가

NAVER_data = fdr.DataReader("035420", "2023-01-01", "2023-10-27")
display(NAVER_data)
```

```python
#NAVER 주가가 20만원 이상일때

NAVER_data_40 = NAVER_data[NAVER_data['Close'] >= 230000].sort_values(by=['Close'])
print(NAVER_data_40)

#df[df['TSLA'] >= 100].sort_values(by=['TSLA'])
```

```python
#NAVER 주가가 17.9만원 이하일때
NAVER_data_17 = NAVER_data[NAVER_data['Close'] <= 179000].sort_values(by=['Close'])
print(NAVER_data_17)
```

```python
#NAVER 주가가 188600 이상이고, 21만원 이하일때
NAVER_data_18 = NAVER_data[(NAVER_data['Close'] >= 188600) & (NAVER_data['Close'] <= 210000)].sort_values(by=['Close'])
print(NAVER_data_18)
```
